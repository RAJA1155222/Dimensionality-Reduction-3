Q8. Real-world applications of eigen decomposition

PCA (Principal Component Analysis): Dimensionality reduction using eigenvectors of covariance matrix.

Vibration analysis / physics: Natural modes of structures correspond to eigenvectors.

Markov chains: Stationary distributions from eigenvectors of transition matrices.

Quantum mechanics: Observables represented as matrices; eigenvectors/eigenvalues correspond to measurable states.

Graph analysis: Spectral clustering uses eigenvectors of graph Laplacian.

Q9. Can a matrix have more than one set of eigenvectors and eigenvalues?

Eigenvalues: Unique (up to multiplicity) for a given matrix.

Eigenvectors: Not unique; any scalar multiple of an eigenvector is also an eigenvector.

Matrices with repeated eigenvalues can have multiple independent eigenvectors forming an eigenspace.

Q10. Eigen-Decomposition in data analysis and ML

PCA: Reduce dimensionality by projecting data along eigenvectors of covariance matrix.

LDA (Linear Discriminant Analysis): Uses eigenvectors of scatter matrices to maximize class separation.

Spectral Clustering: Uses eigenvectors of Laplacian matrix for clustering in graphs.

Recommendation systems: Eigen-decomposition in SVD for matrix factorization.
